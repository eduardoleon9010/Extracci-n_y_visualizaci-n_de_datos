**1. Funci√≥n para crear un gr√°fico** ‚úçÔ∏èüìä

Primero, vamos a definir una funci√≥n que se encargue de crear el gr√°fico para los precios de las acciones. Esta funci√≥n recibir√° como par√°metros los datos de las acciones y el nombre de la compa√±√≠a.

````python
# Funci√≥n para crear el gr√°fico
def grafico_acciones(data, nombre_empresa):
    # Verificar que el DataFrame tenga datos
    if data.empty:
        print(f"No hay datos disponibles para {nombre_empresa}.")
        return

    # Asegurarse de que la columna 'Date' sea de tipo datetime
    data['Date'] = pd.to_datetime(data.index)  # Usamos el √≠ndice que contiene las fechas

    # Crear el gr√°fico con Plotly
    fig = go.Figure()

    # Agregar la serie de precios de cierre
    fig.add_trace(go.Scatter(x=data['Date'], y=data['Close'], mode='lines', name=f'{nombre_empresa} Stock Price'))

    # Personalizar el gr√°fico
    fig.update_layout(
        title=f"{nombre_empresa} Stock Price",
        xaxis_title="Fecha",
        yaxis_title="Precio de Cierre ($)",
        template="plotly_dark"  # Puedes cambiar el tema si lo prefieres
    )
````

**2. Uso yfinance para extraer datos de acciones** üìâ

Ahora vamos a usar yfinance para obtener los datos de las acciones de Tesla y GameStop . Para cada una de las compa√±√≠as, descargamos los datos para los √∫ltimos 365 d√≠as.

````python
# Extraer datos de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  # √öltimo a√±o de datos
````

````python
# Extraer datos de GameStop
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="1y")  # √öltimo a√±o de datos
````

**3. Descargar los datos del √∫ltimo a√±o**

````python
# Descargar datos de Tesla (TSLA) para el √∫ltimo a√±o
tesla_stock = yf.download('TSLA', period='1y')  # √öltimo a√±o
````

````python
# Mostrar los primeros registros
tesla_stock.head()
````

````python
# Descargar datos de GameStop (GME) para el √∫ltimo a√±o
gamestop_stock = yf.download('GME', period='1y')  # √öltimo a√±o
````

````python
# Mostrar los primeros registros
gamestop_stock.head()
````

**4. Convertir los datos a formato HTML**

````python
# Convertir los DataFrames a tablas HTML
tesla_html = tesla_stock.to_html()

# Imprimir las primeras filas de la tabla HTML para Tesla
print("Tabla de Tesla en formato HTML:")
print(tesla_html[:500])  # Imprime solo una parte de la tabla para ver c√≥mo se ve
````

````python
# Convertir los DataFrames a tablas HTML
gamestop_html = gamestop_stock.to_html()

print("Tabla de GameStop en formato HTML:")
print(gamestop_html[:500])  # Imprime solo una parte de la tabla para ver c√≥mo se ve
````

**5. Usar BeautifulSoup para analizar la tabla HTML**

````python
# Convertir las tablas HTML a objetos BeautifulSoup
soup_tesla = BeautifulSoup(tesla_html, 'html.parser')

# Extraer las filas de las tablas
tesla_rows = soup_tesla.find_all('tr')

# Imprimir algunas filas de la tabla de Tesla
for row in tesla_rows[:5]:  # Mostrar las primeras 5 filas
    print(row.text)
````

````python
# Convertir las tablas HTML a objetos BeautifulSoup
soup_gamestop = BeautifulSoup(gamestop_html, 'html.parser')

# Extraer las filas de las tablas
gamestop_rows = soup_gamestop.find_all('tr')

# Imprimir algunas filas de la tabla de GameStop
for row in gamestop_rows[:5]:  # Mostrar las primeras 5 filas
    print(row.text)
````

**üìäüíπ 6. Graficas de las acciones de Tesla Y Gamesstop**

````python
# Llamar a la funci√≥n para graficar los datos de Tesla
grafico_acciones(tesla_data, "Tesla")
````

````python
# Llamar a la funci√≥n para graficar los datos de GameStop
grafico_acciones(gamestop_data, "GameStop")
````
